```python
%%sql

CREATE TABLE newPatients (
    Pregnancies INT,
    Glucose INT,
    BloodPressure INT,
    SkinThickness INT,
    Insulin INT,
    BMI DECIMAL(8,2),
    DiabetesPedigreeFunction DECIMAL(8,2),
    Age INT,
    Outcome INT    
                            );
```

```python
%%sql
INSERT INTO newPatients(Pregnancies,
    Glucose,
    BloodPressure,
    SkinThickness,
    Insulin,
    BMI,
    DiabetesPedigreeFunction,
    Age,
    Outcome)

SELECT  Pregnancies,
    Glucose,
    BloodPressure,
    SkinThickness,
    Insulin,
    BMI,
    DiabetesPedigreeFunction,
    Age,
    Outcome FROM patients WHERE Age > 50;
```

```python
%%sql
SELECT * FROM newPatients
```

```python
%%sql
ALTER TABLE newPatients
ADD perfil Varchar(20)
```

```python
%%sql
UPDATE newPatients
SET perfil = 'Normal'
WHERE BMI < 30
```

```python
%%sql
SELECT * FROM newPatients
```

```python
%%sql
UPDATE newPatients
SET perfil = 'Obese'
WHERE BMI >= 30
```
